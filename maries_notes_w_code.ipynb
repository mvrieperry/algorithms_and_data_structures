```python
import numpy as np
import timeit
import random

size = 10**4 # 10,000

# Using Python lists
list_a = list(range(size)) # [0, 1, 2, 3, ..., 9999]
list_b = list(range(size)) # [0, 1, 2, 3, ..., 9999]
#                            [0, 2, 4, 6, ..., 19998]

# Using NumPy arrays
array_a = np.arange(size) # np.arrays([0, 1, 2, 3, ..., 9999])
array_b = np.arange(size) # np.arrays([0, 1, 2, 3, ..., 9999])

# Timing for list addition
list_time = timeit.timeit(lambda: 
  [a + b for a, b in zip(list_a, list_b)], number=1)

# Timing for vectorized array addition
array_time = timeit.timeit(lambda: 
  array_a + array_b, number = 1)

print(f"List Addition: {list_time:.6f} seconds")
print(f"Vectorized Addition: {array_time:.6f} seconds")
```

```python
# Python list, have duplicates and has an order (so you can ask for the first element, last element etc for instance)
list_time = timeit.timeit(lambda:
  -1 in list_a, number = 1)
  
# Python set, cant store duplicates. And there is no order
set_a = set(range(size)) # This is a set with 1 million elements 
set_time = timeit.timeit(lambda:
  -1 in set_a, number = 1)

print(f"List Search: {list_time:.6f} seconds")
print(f"Set Search: {set_time:.6f} seconds")
```
